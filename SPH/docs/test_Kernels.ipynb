```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo generado por C++
df = pd.read_csv(
    "../test_results/kernel_test.output",
    sep='\s+',   # separador por espacios
    skiprows=1,              # saltar la cabecera (ya la definimos abajo)
    names=["r", "W", "dWx", "dWy"]
)

# Graficar W(r) y su derivada en x
plt.figure(figsize=(8,5))
plt.plot(df["r"], df["W"], label="W(r) - kernel", linewidth=2)
plt.plot(df["r"], df["dWx"], label="dW/dr (proyeccin x)", linestyle="--", linewidth=2)

# Opcional: incluir dWy si quieres ver la otra proyeccin
plt.plot(df["r"], df["dWy"], label="dW/dr (proyeccin y)", linestyle=":")

plt.xlabel("r")
plt.ylabel("Valor")
plt.title("Kernel cbico y derivadas")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
